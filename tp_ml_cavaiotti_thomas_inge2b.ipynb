```python
from sklearn.model_selection import train_test_split

X = df.drop("smoking", axis=1)
y = df["smoking"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X.count()

#Par convention, on utilise 20% du dataset pour le set de test et 80% pour le set d'entrainement.
#C'est une norme largement appliquee car elle permet a l'algorithme de bien s'entrainer sur une base de donnees dense, et de le tester en assurant des performances optimales

#Diviser le set de donnees permet une bonne repartition entre le training set et le test set.
#Sans training set, le systeme ne comprend pas ce qu'il a classifier (ici) et sans test set nous ne pouvons pas savoir si le systeme fonctionne efficacement
```

```python
#modele Regression Logistique

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(max_iter=1000) #on est sur d'avoir le maximum d'iteration
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(f"Precision : {acc*100} %")

#Temps d'execution : 10 secondes
#Resultat obtenu : 70.4 %
```

```python
#modele DNN

#!pip install keras==2.12.0 #Nécessaire en cas de probleme avec 'keras.wrappers', non-present dans les versions ulterieurs

from sklearn.model_selection import GridSearchCV
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
from keras.wrappers.scikit_learn import KerasRegressor

def DNN():
  #Preparation modele DNN
  n_features = X.shape[1]

  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape=(n_features,)))
  model.add(Dense(8, activation='relu'))
  model.add(Dense(1))

  optimizer = RMSprop(0.01)
  model.compile(loss='mean_squared_error',optimizer=optimizer)
  return model

model = KerasRegressor(build_fn=DNN, verbose=1)

#Grille des hyperparametres
batch_size = [10, 20, 40, 60, 80, 100] #Nombre echantillons pour la mise a jour des poids du reseau de neuronnes
epochs = [10, 50, 100] #Nombre de passage sur les donnees d'entrainement
param_grid = dict(batch_size=batch_size, epochs=epochs)

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1)
grid_result = grid.fit(X_train, y_train, verbose=2)

print("Meilleur score : %f utilisant les hyperparametres %s" % (grid_result.best_score_, grid_result.best_params_))

#Temps d'execution : 1 heures (il fallait s'y attendre)
#Meilleur resultat obtenu : -23% / batch_size : 50, epochs : 100
```

```python
#modele Random Forest

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

#Grille des hyperparametres
param_grid = {
'n_estimators': [100, 200], #Nombre d'estimateur: augmente precision et temps de calcul
'max_features': [0.5, 'sqrt'], #Caracteristiques maximales : limite les caracteristiques randant les arbres uniques
'max_depth': [10, 20, None] #Profondeur maximale : eviter le surapprentissage
}

grid_rf = GridSearchCV(estimator=rf,
                       param_grid=param_grid,
                       cv=3,
                       scoring='accuracy')

grid_rf.fit(X_train, y_train)

print("Meilleur score : %f utilisant les hyperparametres %s" % (grid_rf.best_score_, grid_rf.best_params_))

#Temps d'execution : 12 minutes
#Meilleur resultat obtenu : 80.6 % / max_depth: none, max_features: sqrt, n_estimators: 200
```

```python
#modele Naive Bayes

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import PowerTransformer
import numpy as np
from sklearn.model_selection import GridSearchCV

cv_method = RepeatedStratifiedKFold(n_splits=5,
                                    n_repeats=3,
                                    random_state=999)

#Grille des hyperparametres
params_NB = {'var_smoothing': np.logspace(0,-9, num=100)} #Lissage de variance : part plus grande variance de toutes les caracteristiques
                                                          #a ajouter aux variances pour la stabilite des calculs

model = GaussianNB()

gs_NB = GridSearchCV(estimator=model,
                     param_grid=params_NB,
                     cv=cv_method,
                     verbose=1,
                     scoring='accuracy')

Data_transformed = PowerTransformer().fit_transform(X_test)
gs_NB.fit(Data_transformed, y_test)

print("Meilleur score : %f utilisant les hyperparametres %s" % (gs_NB.best_score_, gs_NB.best_params_))

#Temps d'execution : 14 secondes
#Meilleur resultat obtenu : 70.9 % / var_smoothing: 1.0
```

```python
#modele XGBoost

import xgboost as xgb

xgb_model = xgb.XGBClassifier()

#Grille des hyperparametres
param_grid = {
'n_estimators': [100, 200], #Nombre d'arbre : augmente le taux d'information saisie par le modele
'learning_rate': [0.01, 0.1], #Taux d'apprentissage : vitesse et precision de l'apprentissage
'max_depth': [3, 6] #Profondeur maximale : eviter le surapprentissage
}

grid_xgb = GridSearchCV(estimator=xgb_model,
                        param_grid=param_grid,
                        cv=3,
                        scoring='accuracy')

grid_xgb.fit(X_train, y_train)

print("Meilleur score : %f utilisant les hyperparametres %s" % (grid_xgb.best_score_, grid_xgb.best_params_))

#Temps d'execution : 31 secondes
#Meilleur resultat obtenu : 77.5 % / learning_rate: 0.1, max_depth: 6, n_estimators: 200
```
